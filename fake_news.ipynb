```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string

```

```python
data_fake  =  pd.read_csv("Fake news.csv", engine='python', on_bad_lines='skip')
data_true  =  pd.read_csv("True news.csv", engine='python', on_bad_lines='skip')
```

```python
data_fake['class'] = 0
data_true['class'] = 1
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_mannual_testing = data_fake.tail(10)
data_fake = data_fake.iloc[:-10]

data_true_mannual_testing = data_true.tail(10)
data_true = data_true.iloc[:-10]
```

```python
data_fake_mannual_testing.head(10)
```

```python
data_true_mannual_testing.head(10)
```

```python
data_merge = pd.concat([data_fake, data_true], axis =0 )
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data= data_merge.drop(['title', 'subject','date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(["index"], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
   text = text.lower()
   text = re.sub('\[.*?\]', '', text)
   text = re.sub("\\W"," ",text)
   text = re.sub('https?://\S+|www\.\S+', '', text)
   text = re.sub('<.*?>+', '', text)
   text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
   text = re.sub('\n', '', text)
   text = re.sub('\w*\d\w*', '', text)
   return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25,)

vecterization = TfidfVectorizer()
xv_train = vecterization.fit_transform(x_train)
xv_test = vecterization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
print(f"Shape of xv_train: {xv_train.shape}")
print(f"Shape of y_train: {y_train.shape}")

LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB = GradientBoostingClassifier(random_state=0)
GB.fit(xv_train, y_train)
```

```python
predit_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test, y_test)
```

```python
pred_dt = DT.predict(xv_test)
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(random_state=0)
RF.fit(xv_train, y_train)
```

```python
pred_rcc = RF.predict(xv_test)
```

```python
RF.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rcc))
```

```python
def output_label(n):
  if n == 0:
    return "Fake News"
  elif n == 1:
    return "Not A Fake News"

def manual_testing(news):
  testing_news = {"text":[news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)
  new_x_test = vecterization.transform(new_def_test["text"])
  pred_lr = LR.predict(new_x_test)
  pred_dt = DT.predict(new_x_test)
  pred_gb = GB.predict(new_x_test)
  pred_rf = RF.predict(new_x_test)
  return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGB Prediction: {} \nRF Prediction: {}".format(output_label(pred_lr[0]),
                                                                                                          output_label(pred_dt[0]),
                                                                                                          output_label(pred_gb[0]),
                                                                                                          output_label(pred_rf[0])))
```
